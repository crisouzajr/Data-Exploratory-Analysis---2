### <font color ='dark_purple_3'> EXTRA - Utilização da Função lambda para criar uma função de coversão de unidade de temperatura.

```python
'''Celcius para Fahrenheit'''

Celsius = [24]

Fahrenheit = map(lambda x: (float(9)/5)*x + 32, Celsius )
    
print(list(Fahrenheit))
```

```python
'''Fahrenheit para Celcius'''

Fahrenheit = [4.1, 80.1]

Celsius = map(lambda x: (float(5)/9)*(x - 32), Fahrenheit )

print(list(Celsius))
```
